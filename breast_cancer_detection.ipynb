## Adding Dependency

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as py
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
```

# Data Collection & Processing

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns = breast_cancer_dataset.feature_names)
```

```python
data_frame['label']=breast_cancer_dataset.target
```

# Separating the features and target

```python
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

## Spilting the data into training data and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_test.shape,x_train.shape)
```

# Model Training 
# Logistic Regression

```python
# Model Training 
# Logistic Regression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

# Model Evaluation 

Accuracy Score

```python
# Accuracy score of train data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print("Accuracy Score of this Training data is..",training_data_accuracy*100,"%")
```

```python
# Accuracy score of test data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(y_test,x_test_prediction)
```

```python
print("Accuracy Score of this Testing data is..",testing_data_accuracy*100,"%")
```
